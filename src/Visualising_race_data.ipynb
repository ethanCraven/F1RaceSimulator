```python

import pandas as pd
import matplotlib.pyplot as plt
```

This is a notebook for visualising previous race data. This is a key component of the research and implementation. While this step would have been completed regardless, the main motivation for implementing this data analysis is that over the last season (2024) there have been many many races where tyre degredation hasn't actually been a massive factor. Teams and drivers have seemingly found the sweet spot for each tyre at each track. This means that often lap time graphs have very consistent times over the same stint, with the second or third stiny naturally being quicker. This will change the tyre degredation model since (possibly) it doesn't need to be as complicated. 

```python
df = pd.read_csv('all_race_data.csv')
```

```python
norris_silverstone_df_2023 = (df[(df.race_name == "British Grand Prix") & (df.driver_id == "norris") & (df.year == 2023)])
print(norris_silverstone_df_2023)
```

```python
plt.figure(figsize=(15,9))
plt.plot(norris_silverstone_df_2023.lap_number, norris_silverstone_df_2023.lap_time)
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (sec)")
plt.title("Lap times for Norris at the 2023 British Grand prix")
```

This is an example of the afore mentioend point. Other than the pitstop and what was obviously a saftey car, there is no evidence here of what was originally thought. While the second stint is quicker than the first due to fuel loads (which was expected) the timings are consistent, all sitting at that 90-92 seconds range. 

```python
russell_spanish_2024_df = (df[(df.race_name == "Spanish Grand Prix") & (df.driver_id == "russell") & (df.year == 2024)])
plt.figure(figsize=(15,9))
plt.plot(russell_spanish_2024_df.lap_number, russell_spanish_2024_df.lap_time)
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (sec)")
plt.title("Lap times for Russell at the 2024 Spanish Grand prix")
```

Here, a better trend can be observed. This is a 2 stop race, since there were two pit stops, it may be found that 2 stop races are much better at showing an increase in lap time over the stint. This is more evidence to show that drivers and teams have found the sweet spot within the tyre. This conclusion is due to the differences in comparison, there is a more noticable increase in time gain due to degredation in a two stop race because the drivers don't have to conserve for a longer duration. The reason this is important is because it shifts the focus of the simulation away from just the standard 'the tyre gets older so the car gets proportionally slower' mindset and towards a mindset that is driven by driver ability and application of pace. For example, two pit stops and going at 100% pace might end up being slower than one pit stop and a more conservative pace. This is definitely a decision that the model will have to make.

Following on from this, the next study will be to see how different tracks have different degredation curves and in turn sometimes a different number of pit stops. This should give indications on how to make the tyre degredation curve modular for each track. To do this the aim is to look at all the tracks on the calendar, and compare the number of corners, distance to the average change in lap time lap by lap. 

Most people would assume that as the race goes on the times get slower due to older tyres, but as you can see they actually get quicker on average by just shy of a tenth each lap. Let's see if this trend continues.

```python
driver_race_2024_df = (df[(df.race_name == "Dutch Grand Prix") & (df.driver_id == "norris") & (df.year == 2024)])
total_time = 0
time_changes = []
prev_lap = driver_race_2024_df.iloc[0, 5]
for index, row in driver_race_2024_df.iterrows():
    total_time += row['lap_time']
    time_change = prev_lap - row['lap_time']
    time_changes.append(time_change)
    prev_lap = row['lap_time']
total_time_change = sum(time_changes[2:])
print(f"Total time change: {total_time_change}")
print(time_changes[2:])
avg_time_change = sum(time_changes[2:]) / (len(time_changes))
print(f"Average time change: {avg_time_change}")
```
