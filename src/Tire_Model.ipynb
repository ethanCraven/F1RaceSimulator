```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulated data for tire age and lap times
tire_age = np.linspace(0, 50, 100)  # Tire age in laps (0 to 50 laps)
coefficients = {
    # [0, 1, 2]
    "soft": [2.3, 0.0025, 0.0095],
    "medium": [3.0, 0.0015, 0.004],
    "hard": [4, 0.002, 0.0019]
}

lap_times ={}
for compound, (beta0, beta1, beta2) in coefficients.items():
    lap_times[compound] = beta0 + beta1 *tire_age + beta2 * (tire_age ** 2)

plt.figure(figsize=(15, 9))
for compound, times in lap_times.items():
    plt.plot(tire_age, times, label=f"{compound.capitalize()}")

plt.title("Tire Model: Influence of Tire Age on Lap Times")
plt.xlabel("Tire Age (laps)")
plt.ylabel("Residual Lap Time (seconds)")
plt.legend(title="Tire Compound")
plt.grid(True)
plt.show()
```

The code below is the first idea for the improved tyre graph using the soft compound tyre. As you can see the first 6 laps are able to be quicker with the rest being slower and at lap number 17 maybe, is time for an optimal pitstop. Probably a bit long of a stint for the soft tyres maybe. but we will see. 

```python
total_time_soft = 0 
for i in range(1, 57):
    tyre_affect = 2+0.002*i + 0.007 * (i **2)
    fuel_affect = i*0.057
    lap_time = 90 + tyre_affect - fuel_affect
    print(f"Lap number: {i} Time: {lap_time}")
    total_time_soft = total_time_soft + lap_time
print(f"Total time for this compound = {total_time_soft} seconds")
```

Next the project looks at the medium tyres and how they should degrade over time.

```python
total_time_medium = 0
for i in range(1, 57):
    tyre_affect = 4+0.0015*i + 0.004 * (i **2)
    fuel_affect = i*0.057
    lap_time = 90 + tyre_affect - fuel_affect
    print(f"Lap number: {i} Time: {lap_time}")
    total_time_medium = total_time_medium + lap_time
print(f"Total time:  {total_time_medium}")
```

Finally, the same is done for the hard tyres.

```python
total_time_hard = 0
for i in range(1, 57):
    tyre_affect = 6+0.002*i + 0.0019 * (i **2)
    fuel_affect = i*0.057
    lap_time = 90 + tyre_affect - fuel_affect
    print(f"Lap number: {i} Time: {lap_time}")
    total_time_hard = total_time_hard + lap_time
print(f"Total time:  {total_time_hard}")
```
